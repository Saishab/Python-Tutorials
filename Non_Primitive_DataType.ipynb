##### Unpacking Tuples
Unpacking means assigning each element of a tuple to a variable.

```python
student = ("Dhirendra", 15, "Science")

name, age, program = student

print(name)
print(age)
print(program)
```

##### Set Operations

###### Union – Combines all unique elements

```python
A = {1, 2, 3}
B = {3, 4, 5}
print(A.union(B))   
print(A | B) 
```

###### Intersection – Only common elements

```python
A = {1, 2, 3}
B = {2, 3, 4}
print(A.intersection(B))  
print(A & B)
```

###### Difference – Items present in A but not in B

```python
A = {1, 2, 3}
B = {2, 3, 4}
print(A.difference(B)) 
print(A - B)
```

###### Symmetric Difference

```python
# Elements that are in A or B but not in both.

A = {1, 2, 3}
B = {3, 4, 5}
print(A.symmetric_difference(B))
print(A ^ B)
```
