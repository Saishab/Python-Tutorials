#### While Loop 

![Election](election.png){width=50 height=140}


```python

ballot_papers = 10   # hypothetical

while ballot_papers > 0:
    print("Counting a ballot paper..")
    ballot_papers = ballot_papers - 1  # removing counted ballot paper ( 1 at a time)  from the box
    print("The remaining papers are",ballot_papers)
    

print("Ballot box is empty")
```

```python
### queue in voting center 

queue = 4

while queue > 0:
    print("One voter has voted")
    queue -= 1  # queue = queue - 1 

print("Polling station closed")
```

```python
votes = 7

while votes > 0:
    print("Counting vote...")
    votes -= 1 #votes = votes - 1
    print("Remaining votes",votes)

print("All votes counted")
```

##### Nested Loop

```python
for party in range(1, 5):  
    print("Party", party) 
    for candidate in range(1, 5):  
        print(" Candidate", candidate)
        
# 140 Political Parties 1,2,...,140
# 165 election areas = 140*165 ,1[1,...165],2[1,...165]
```

```python
## Votes per Candidate per Booth
booths = 3
candidates = 2

for booth in range(1, booths + 1): 
    print(f"\nPolling Booth {booth}")

    for candidate in range(1, candidates + 1):
        votes = int(input(f" Enter predicted votes for Candidate {candidate}: "))
        print(f" Candidate {candidate} received {votes} votes")
```

#### Continue 

```python
voters = ["Nepali", "Nepali", "Nepali", "Nepali",
          'USA-GC','Germany- Blue Card','Nepali']

count = 0   # to count Nepali voters

for person in voters:
    if person != "Nepali":
        continue
    count += 1

print("Total Nepali voters:", count)

```
