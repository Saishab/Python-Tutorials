### When to use conditional statements?

```python
## exam result 

physics_marks = int(input("Enter student marks in Physics"))

if physics_marks >= 32:
    print("You are pass")

else:
    print("Fail vais vai")

# elif -- helps in multiple conditional statements  
```

```python
## identation case
if physics_marks > 40: print( "You are pass, badhai chh") 
```

```python
## else what? 

physics_marks = int(input("Enter your marks in physics")) 

if physics_marks > 40: 
    print("You are pass") 
elif physics_marks < 20: 
    print("Dherai mehenat garna parne dekhiyo kanchha") 
else:
    print("Result leko chhaina") 
```

```python
marks_in_math = float(input("Enter your marks in Mathematics")) 

if marks_in_math >= 80:
    print("You scored Grade A") 
    
elif marks_in_math >= 70 and marks_in_math < 80: 
    print("You scored Grade B") 
    
elif marks_in_math >= 60 and marks_in_math < 70: 
    print("You scored Grade C")
    
elif marks_in_math < 60: 
    print("You scored Grade D")

else: 
    print("Enter valid marks") 
```

```python
## check if a number is even or odd 

x = int(input("Enter a number"))

if x % 2 == 0: 
    print("Even") 
else: 
    print("Odd") 
```

```python
# first creation of variable [dictionary] to store details of constituent areas of Nepal [taking only valley]
constituent_areas = {
    "kathmandu": 13,
    "bhaktapur": 2,
    "lalitpur": 4
}

# Ask if the user is Nepali
nepali = input("Are you Nepali? (yes/no): ").lower()

if nepali == "yes":
    # asking if user is in aboard
    location = input("Are you in Nepal or abroad? ").lower()
    
    if location == "nepal":
        print("You are inside Nepal. Here is the total count of constituent areas in each district inside valley:")
        print("Kathmandu:", constituent_areas["kathmandu"], "areas")
        print("Bhaktapur:", constituent_areas["bhaktapur"], "areas")
        print("Lalitpur:", constituent_areas["lalitpur"], "areas")
    else:
        print("You cannot proceed for constituency selection as you are abroad.")
else:
    print("You cannot vote in Election 2082.")
```

```python
###### multi conditions 

score = int(input("Enter a marks"))

grade = "A" if score >= 90 else \     # '\' is the syntax needed for multi condition
        "B" if score >= 80 else \
        "C" if score >= 70 else \
        "F" 
print(grade) ## function-- same variable or function will be used in multiple palces 
```
