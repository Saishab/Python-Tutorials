```python
## identifying the numbers first and then asking to enter ther name of parties 

parties = []   

n = int(input("Enter number of parties: "))  

for i in range(n):    
    party = input("Enter party name: ")   
    parties.append(party)


for party in parties: 
    print(party, "is participating in Election 2082")  
```

```python
parties = []

n = int(input("Enter number of parties: "))

for i in range(1, n + 1):
    party = input(f"Enter {i} party name: \n")
    parties.append(party)

for party in parties:
    print(party, "is participating in Election 2082")
```

#### While Loop 

![Election](election.png){width=50 height=140}


```python

ballot_papers = 10   # hypothetical

while ballot_papers > 0:
    print("Counting a ballot paper..")
    ballot_papers = ballot_papers - 1  # removing counted ballot paper ( 1 at a time)  from the box


print("Ballot box is empty")
```

```python
#password checking 
password = ""

while password != "vote2082":
    password = input("Enter election commission admin password: ")

print("Access granted to Election System ")
```

```python
### queue in voting center 

queue = 4

while queue > 0:
    print("One voter has voted")
    queue -= 1

print("Polling station closed")
```

```python
votes = 7

while votes > 0:
    print("Counting vote...")
    votes -= 1
    print("Remaining votes",votes)

print("All votes counted")
```

##### Nested Loop

```python
for party in range(1, 3): 
    print("Party", party) 
    for candidate in range(1, 3): 
        print(" Candidate", candidate)
```

```python
## Votes per Candidate per Booth
booths = 3
candidates = 2

for booth in range(1, booths + 1):
    print(f"\nPolling Booth {booth}")

    for candidate in range(1, candidates + 1):
        votes = int(input(f" Enter predicted votes for Candidate {candidate}: "))
        print(f" Candidate {candidate} received {votes} votes")
```
